# Hardware GPT

This GPT will be trained on computer hardware, their functions, and be able to answer questions based on them.

**Some questions/answers are beyond the desktop and internal components for necessary context.**

```python
import gpt_2_simple as gpt2

session = gpt2.start_tf_sess()
```

```python
gpt2.finetune(
    session, 
	"hardware.txt", 
	model_name="124M", 
	steps=100, 
	run_name="hardware"
)
```

```python
gpt2.generate(session, prefix="Q: What is inside of a computer?", run_name="hardware")

gpt2.generate(session, prefix="Q: What is the purpose of the power supply?", run_name="hardware")

gpt2.generate(session, prefix="Q: Is using a dedicated or intergrated graphics card better?", run_name="hardware")
```
